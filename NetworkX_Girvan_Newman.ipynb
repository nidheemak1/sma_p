# Practical 25 - Analyze Zachary's Karate Club network using NetworkX. Detect communities using the Girvan‚ÄìNewman algorithm and interpret the splits.


```python
import networkx as nx
import matplotlib.pyplot as plt
from networkx.algorithms.community import girvan_newman
import itertools
```

```python
G = nx.karate_club_graph()
```

```python
from networkx.algorithms.community import girvan_newman

communities = girvan_newman(G)
```

```python
first_level_communities = next(communities)
comm_list = [list(c) for c in first_level_communities]
```

```python
print("üîç Detected Communities (First Split):")
for i, group in enumerate(comm_list):
    print(f"Community {i + 1}: {group}")
```

```python
color_map = {}
for i, community in enumerate(comm_list):
    for node in community:
        color_map[node] = i

node_colors = [color_map[node] for node in G.nodes()]

plt.figure(figsize=(10, 7))
pos = nx.spring_layout(G, seed=42)
nx.draw(G, pos, node_color=node_colors, with_labels=True, cmap=plt.cm.Set1, node_size=800)
plt.title("Zachary's Karate Club - Girvan‚ÄìNewman Community Detection (First Split)")
plt.show()
```
